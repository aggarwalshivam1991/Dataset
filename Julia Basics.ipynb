```julia
using ScikitLearn, CSV
```

```julia
iris = CSV.read("iris.csv",normalizenames = true)
```

```julia
#Machine leaning algo and libraries only take arrays
```

```julia
iris_x =convert(Array,iris[:,[1,2,3,4]])
```

```julia
iris_y = convert(Array,iris[:,5])
```

```julia
@sk_import linear_model: LogisticRegression
```

```julia
log_reg_model = LogisticRegression()
#LogisticRegression is a machine learning alogorithm
```

```julia
#above is not a error it says that everything is fine showing the elements
```

```julia
fit!(log_reg_model,iris_x,iris_y)
```

```julia
predictions = predict(log_reg_model,iris_x)
```

```julia
@sk_import metrics: accuracy_score
# it is a library in scikitlearn to know the accuracy of the result. the above shows setosa is on top and virginica at 2
```

```julia
accuracy = accuracy_score(predictions,iris_y)
```

```julia
#above dataset is 97.33 % accurate. Lets' move on to tree model
```

```julia
@sk_import tree: DecisionTreeClassifier
```

```julia
tree_model = DecisionTreeClassifier()
```

```julia
fit!(tree_model,iris_x,iris_y)
```

```julia
tree_predictions = predict(tree_model,iris_x)
```

```julia
tree_accuracy = accuracy_score(tree_predictions,iris_y)
```

```julia
@sk_import ensemble: RandomForestClassifier
```

```julia
rf_model = RandomForestClassifier()
```

```julia
fit!(rf_model,iris_x,iris_y)
```

```julia
rf_predictions = predict(rf_model,iris_x)
```

```julia
rf_accuracy = accuracy_score(rf_predictions,iris_y)
```

```julia
#1 means 100% accuracy
```

```julia
@sk_import model_selection: train_test_split
```

```julia
x_train,x_test,y_train,y_test = train_test_split(iris_x,iris_y,random_state=1)
```

```julia
log_reg_2 = LogisticRegression()
```

```julia
fit!(log_reg_2,x_train,y_train)
```

```julia
prediction_log_reg = predict(log_reg_2,x_test)
```

```julia
acuracy_log_reg= accuracy_score(prediction_log_reg,y_test)
```

```julia
#this is 97.63 accuracy using Train & test machine learning technique
```
